```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [8,6]
print("Setup is completed")
```

## Anh 1

```python
image = cv2.imread("./homework-data/01.jpg")
#convert our image to gray
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#our kernel
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (12,12))
#Show the original image
plt.imshow(image[:,:,::-1])
```

```python
#show gray picture
plt.imshow(gray, cmap='gray')
```

```python
#thresholding
_, thresh = cv2.threshold(gray, 130, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
```

```python
#Dilation
dilate_img = cv2.dilate(thresh, kernel, iterations=4)
plt.imshow(dilate_img, cmap='gray')
```

```python
#Erosion
eros_gray = cv2.erode(dilate_img, kernel, iterations=30)
plt.imshow(eros_gray, cmap='gray')
```

```python
dilate_img = cv2.dilate(eros_gray, kernel, iterations=30)
plt.imshow(dilate_img, cmap='gray')
```

```python
#split area
nlabel,labels,stats,centroids = cv2.connectedComponentsWithStats(dilate_img,connectivity=8)
#quantity of area
print(nlabel) 
```

```python
#choose which area we want to keep, others will be black
mask = np.zeros_like(dilate_img, dtype=np.uint8)
mask[labels==2] = 255
#Finally, print out the result
plt.imshow(mask, cmap='gray')
```

```python
#Find the countours and draw on the original image
contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#print(contours)
img_write = image.copy()
cv2.drawContours(img_write, contours, 0, (0,255,0), 20)
plt.imshow(img_write[:,:,::-1])
cv2.imwrite("01_detect.jpg", img_write)
```
